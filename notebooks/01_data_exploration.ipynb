## Key EDA Insights (Day 2)

- Overall churn rate is ~26.5%, indicating a significant retention problem
- Month-to-month customers show disproportionately high churn
- Customers with higher monthly charges are more likely to churn
- Longer contract duration strongly correlates with lower churn and higher tenure

These findings suggest contract type, tenure, and monthly charges as key drivers for churn prediction.
